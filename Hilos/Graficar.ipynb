```python
# Importamos librerias
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Abrimos el archivo de los resultados en excel
resultados = pd.read_csv('resultadosfourier.csv',header=None)
ejeX = list(range(resultados.shape[0])) 

resultados
```

```python
# Abrimos los archivos generados por C
y1 = pd.read_csv('./y1.csv',header=None)
y2 = pd.read_csv('./y2.csv',header=None)
y3 = pd.read_csv('./y3.csv',header=None)

# Creamos una figura y un objeto que nos permite manipular los grÃ¡ficos de la figura
fig, ax =  plt.subplots(3,2)
# Ajustamos el tamaÃ±o de la figura
fig.set_size_inches(16,12)
# GrÃ¡fico 1
ax[0,0].stem(ejeX, resultados[0],markerfmt='C0o', linefmt='C0-')
# GrÃ¡fico 2
ax[0,1].stem(ejeX, y1,markerfmt='C1o', linefmt='C1-')
# GrÃ¡fico 3
ax[1,0].stem(ejeX, resultados[1],markerfmt='C0o', linefmt='C0-')
# GrÃ¡fico 4
ax[1,1].stem(ejeX, y2,markerfmt='C1o', linefmt='C1-')
# GrÃ¡fico 5
ax[2,0].stem(ejeX, resultados[2],markerfmt='C0o', linefmt='C0-')
# GrÃ¡fico 6
ax[2,1].stem(ejeX, y3,markerfmt='C1o', linefmt='C1-')
# Mostramos el grafico
plt.show()
```
